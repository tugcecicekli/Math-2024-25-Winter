### **1. Definition of the Derivative**

The **derivative** of a function $f(x)$ at a point $x = c$ is the limit of the difference quotient as $h \to 0$:

$$f'(c) = \lim_{h \to 0} \frac{f(c+h) - f(c)}{h}$$
If this limit exists, $f(x)$ is said to be **differentiable** at $x = c$.

- The derivative represents the **rate of change** or **slope** of the function at that point.
- It can be interpreted geometrically as the slope of the tangent line to the curve at $x = c$.

---

### **2. Derivative Rules**

#### A. **Constant Rule**
$\frac{d}{dx}[c] = 0$
- The derivative of a constant is zero.

#### B. **Power Rule**
$\frac{d}{dx}[x^n] = n \cdot x^{n-1}$
- For any real number $n$, the derivative of $x^n$ is $n \cdot x^{n-1}$.

#### C. **Constant Multiple Rule**
$\frac{d}{dx}[c \cdot f(x)] = c \cdot f'(x)$
- If a function is multiplied by a constant, the derivative is the constant multiplied by the derivative of the function.

#### D. **Sum and Difference Rule**
$\frac{d}{dx}[f(x) \pm g(x)] = f'(x) \pm g'(x)$
- The derivative of a sum or difference is the sum or difference of the derivatives.

#### E. **Product Rule**
$\frac{d}{dx}[f(x) \cdot g(x)] = f'(x) \cdot g(x) + f(x) \cdot g'(x)$
- The derivative of a product of two functions is given by the derivative of the first times the second plus the first times the derivative of the second.

#### F. **Quotient Rule**
$\frac{d}{dx}\left[\frac{f(x)}{g(x)}\right] = \frac{f'(x) \cdot g(x) - f(x) \cdot g'(x)}{[g(x)]^2}$
- The derivative of a quotient is given by the quotient of the derivatives of the numerator and denominator, with a subtraction in the numerator and the denominator squared.

#### G. **Chain Rule**
$\frac{d}{dx}[f(g(x))] = f'(g(x)) \cdot g'(x)$
- The derivative of a composition of functions is the derivative of the outer function evaluated at the inner function times the derivative of the inner function.

---

### **3. Derivatives of Common Functions**

- $\frac{d}{dx}[x^n] = n \cdot x^{n-1}$
- $\frac{d}{dx}[e^x] = e^x $
- $\frac{d}{dx}[\ln x] = \frac{1}{x}$
- $\frac{d}{dx}[\sin x] = \cos x$
- $\frac{d}{dx}[\cos x] = -\sin x$
- $\frac{d}{dx}[\tan x] = \sec^2 x $
- $\frac{d}{dx}[\arcsin x] = \frac{1}{\sqrt{1 - x^2}} $
- $\frac{d}{dx}[\arccos x] = -\frac{1}{\sqrt{1 - x^2}} $
- $\frac{d}{dx}[\arctan x] = \frac{1}{1 + x^2}$

---

### **4. Higher-Order Derivatives**

- **Second Derivative**: The derivative of the derivative $f''(x) = \frac{d^2}{dx^2} f(x)$.
- **Third Derivative**: The derivative of the second derivative $f^{(3)}(x) = \frac{d^3}{dx^3} f(x)$.
- Higher-order derivatives describe the rate of change of the rate of change (e.g., acceleration as the derivative of velocity).

---

### **5. Implicit Differentiation**

When a function is given implicitly (i.e., not solved explicitly for $y$ in terms of $x$, we can differentiate both sides with respect to $x$, treating $y$ as a function of $x$ (and using the chain rule):
$\frac{d}{dx}[F(x, y)] = 0 \quad \text{implicitly, differentiate with respect to } x.$

For example:

$x^2 + y^2 = 25 \quad \text{implies} \quad 2x + 2y \frac{dy}{dx} = 0.$

### **6. Applications of Derivatives**

#### A. **Finding Tangent Lines**
- The slope of the tangent line to the curve $y = f(x)$ at a point $x = c$ is given by $f'(c)$.  
The equation of the tangent line is:
$y - f(c) = f'(c)(x - c)$

#### B. **Optimization Problems**
- Derivatives are used to find the **maximum** or **minimum** values of a function.
- Steps:
  1. Find $f'(x)$.
  2. Solve $f'(x) = 0$ to find critical points.
  3. Use the second derivative test or other methods to determine if the critical point is a maximum or minimum.

#### C. **Related Rates**
- Derivatives can be used to solve problems involving rates of change of two or more variables that are related.  
Example: The rate of change of the area of a circle when the radius is changing.

---

### **7. Critical Points and Local Extrema**

- **Critical Points**: Points where $f'(x) = 0$ or $f'(x)$ does not exist.
- **Local Maximum**: If $f'(x) = 0$ and $f''(x) < 0$, the function has a local maximum at that point.
- **Local Minimum**: If $f'(x) = 0$ and $f''(x) > 0$, the function has a local minimum at that point.

---

### **8. Concavity and Inflection Points**

- **Concave Up**: If $f''(x) > 0$, the graph of $f(x)$ is concave up (shaped like a "U").
- **Concave Down**: If $f''(x) < 0$, the graph of $f(x)$ is concave down (shaped like an "n").
- **Inflection Point**: A point where the concavity changes (i.e., $f''(x) = 0$ and changes sign).

---

### **9. Mean Value Theorem**

- The Mean Value Theorem states that for a continuous and differentiable function $f(x)$ on the interval $[a, b]$, there exists at least one point $c \in (a, b)$ such that:
$f'(c) = \frac{f(b) - f(a)}{b - a}$
- This means there is a point where the instantaneous rate of change equals the average rate of change over the interval.

---

### **10. L’Hopital’s Rule (for Indeterminate Forms)**

- When you encounter an indeterminate form like $\frac{0}{0} or \frac{\infty}{\infty}$, you can apply L'Hopital’s Rule:
$$
\lim_{x \to c} \frac{f(x)}{g(x)} = \lim_{x \to c} \frac{f'(x)}{g'(x)} \quad \text{if the limit on the right-hand side exists}.
$$
