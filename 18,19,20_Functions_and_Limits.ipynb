# **Functions**

---

### **1. Definition of a Function**
A function is a relation between two sets where each element in the first set (domain) is associated with exactly one element in the second set (range).  

- **Notation**: $(f: A\to B)$, where $A$ is the domain and $B$ is the codomain.  
- If $y = f(x)$, $x$ is the input, and $y$ is the output.  

---

### **2. Key Terms**

1. **Domain**: The set of all possible input values $(x)$ for which the function is defined.  
2. **Range**: The set of all actual output values $(f(x))$.  
3. **Codomain**: The set of all potential outputs, which may include values not in the range.  
4. **Image**: The output $f(x)$ corresponding to a specific input $x$.  

---

### **3. Types of Functions**

1. **Based on Relation**:
   - **One-to-One (Injective)**: Each element of the domain maps to a unique element of the codomain.  
   - **Onto (Surjective)**: Every element of the codomain is the image of at least one element from the domain.  
   - **Bijective**: Both injective and surjective (one-to-one and onto).

2. **Based on Algebraic Nature**:
   - **Polynomial Function**: $f(x) = a_nx^n + a_{n-1}x^{n-1} + \dots + a_0$ e.g., $f(x) = x^2 + 3x + 2$.  
   - **Rational Function**: $ f(x) = \frac{P(x)}{Q(x)}$, where $P(x)$ and $Q(x)$ are polynomials.  
   - **Exponential Function**: $f(x) = a^x$, where $a > 0$ and $a \neq 1$.  
   - **Logarithmic Function**: $f(x) = \log_a(x)$, the inverse of the exponential function.

3. **Based on Behavior**:
   - **Linear Function**: $f(x) = mx + c$.  
   - **Quadratic Function**: $f(x) = ax^2 + bx + c$.  
   - **Periodic Function**: Repeats values at regular intervals e.g., $( \sin(x) ), ( \cos(x) )$.  
   - **Piecewise Function**: Defined by different expressions in different intervals.  

4. **Based on Input and Output**:
   - **Real-Valued Function**: Domain and range are subsets of real numbers $(\mathbb{R})$.  
   - **Complex Function**: Involves complex numbers $(\mathbb{C})$.  

---

### **4. Representation of Functions**

1. **Algebraic Form**: Using an equation e.g., $(f(x)= x^2 )$.  
2. **Graphical Form**: A visual representation on a coordinate plane.  
3. **Tabular Form**: A table of input-output pairs.  
4. **Mapping Diagram**: Arrows showing input-output associations.  

---

### **5. Operations on Functions**

1. **Addition**: $((f + g)(x)) = f(x) + g(x)$.  
2. **Subtraction**:$((f - g)(x)) = f(x) - g(x)$.  
3. **Multiplication**: $( (f \cdot g)(x)) = f(x) \cdot g(x) )$.  
4. **Division**: $( \left(\frac{f}{g}\right)(x) = \frac{f(x)}{g(x)}),( g(x) \neq 0 )$.  
5. **Composition**: $( (f \circ g)(x)) = f(g(x))$.  

---

### **6. Special Types of Functions**

1. **Identity Function**: $f(x) = x$.  
2. **Constant Function**: $f(x) = c$, where $c$ is constant.  
3. **Even Function**: Symmetric about the y-axis; $f(-x)=f(x)$.  
4. **Odd Function**: Symmetric about the origin; $f(-x)=-f(x)$.  
5. **Inverse Function**: If $f(x) = y$, then $f^{-1}(y)=x$.  
6. **Step Function**: A function that increases or decreases in steps (e.g., floor or ceiling functions).

---

### **7. Graphical Features**

1. **Intercepts**:
   - $x$-Intercept: Where $f(x) = 0$.  
   - $y$-Intercept: Where $x = 0$.  

2. **Symmetry**:
   - Symmetric about the y-axis: Even functions.  
   - Symmetric about the origin: Odd functions.

3. **Asymptotes**:
   - Vertical: $x = a$, where $(f(x) \to \infty )$ or $(f(x) \to -\infty )$ as $(x \to a)$.  
   - Horizontal: $y = b$, where $(f(x) \to b)$ as $(x \to \infty)$ or $(x \to -\infty)$.  

4. **Behavior**:
   - Increasing: $f'(x) > 0$.  
   - Decreasing: $f'(x) < 0$.

#**Limits of Sequences**
---

### **1. Definition of the Limit of a Sequence**
A sequence ${a_n}$ is said to **converge** to a limit $L$ if, for every $\epsilon > 0$, there exists a positive integer $N$ such that:  
$|a_n - L| < \epsilon \quad \text{for all } n \geq N$
- $L$: The limit of the sequence.
- Denoted as:  
  $\lim_{n \to \infty} a_n = L$

If no such $L$ exists, the sequence is said to **diverge**.

---

### **2. Types of Convergence**
#### A. **Finite Limit**
- A sequence converges to a finite value $L$.
- Example:
  $a_n = \frac{1}{n}, \quad \lim_{n \to \infty} a_n = 0.$

#### B. **Infinite Limit**
- A sequence diverges to $\infty$ or $-\infty$.
- Example:
$a_n = n^2, \quad \lim_{n \to \infty} a_n = \infty.$

#### C. **Divergence**
- The sequence does not approach a specific value.
- Example:
  $a_n = (-1)^n, \quad \text{oscillates between } -1 \text{ and } 1.$

---

### **3. Rules for Finding Limits**

#### A. **Limit of a Constant Sequence**
$\lim_{n \to \infty} c = c, \quad \text{where } c \text{ is a constant.}$
- Example: $a_n = 5$, $\lim_{n \to \infty} a_n = 5 $.

#### B. **Sum/Difference Rule**
$\lim_{n \to \infty} (a_n \pm b_n) = \lim_{n \to \infty} a_n \pm \lim_{n \to \infty} b_n.$

#### C. **Product Rule**
$\lim_{n \to \infty} (a_n \cdot b_n) = (\lim_{n \to \infty} a_n) \cdot (\lim_{n \to \infty} b_n)$.

#### D. **Quotient Rule**
$\lim_{n \to \infty} \frac{a_n}{b_n} = \frac{\lim_{n \to \infty} a_n}{\lim_{n \to \infty} b_n}, \quad \text{if } \lim_{n \to \infty} b_n \neq 0.$

#### E. **Power Rule**
$\lim_{n \to \infty} (a_n)^k = (\lim_{n \to \infty} a_n)^k, \quad \text{if } k \text{ is a positive integer.}$

---

### **4. Special Limits**

#### A. **Harmonic Sequence**
$\lim_{n \to \infty} \frac{1}{n} = 0$.

#### B. **Geometric Sequence**
- If $|r| < 1$:  
  $\lim_{n \to \infty} r^n = 0.$
- If $|r| \geq 1$: The sequence diverges.

#### C. **Exponential Growth**
$\lim_{n \to \infty} a^n =
\begin{cases}
0 & \text{if } 0 < a < 1, \\
\infty & \text{if } a > 1.
\end{cases}$

#### D. **Logarithmic Sequence**
$\lim_{n \to \infty} \frac{\ln n}{n} = 0.$

#### E. **Factorial Sequence**
$\lim_{n \to \infty} \frac{n!}{n^n} = 0.$

---

### **5. Squeeze Theorem**
If $a_n \leq b_n \leq c_n$ for all $n \geq N$, and:  
$\lim_{n \to \infty} a_n = \lim_{n \to \infty} c_n = L,$
then:  
$\lim_{n \to \infty} b_n = L.$
- Example:
  $a_n = \frac{\sin n}{n}, \quad -\frac{1}{n} \leq \frac{\sin n}{n} \leq \frac{1}{n}$
  Using the Squeeze Theorem: $\lim_{n \to \infty} \frac{\sin n}{n} = 0$.

---

### **6. Divergent Sequences**
A sequence diverges if:
- $a_n \to \infty$ or $a_n \to -\infty$.
- The terms oscillate without approaching a fixed value.

Example:  
$a_n = (-1)^n, \quad \text{oscillates between } -1 \text{ and } 1$.

---

### **7. Useful Techniques for Finding Limits**

#### A. **Divide by Dominant Term**
For sequences involving polynomials or ratios:  
- Divide numerator and denominator by the highest power of $n$.
- Example:  
  $a_n = \frac{2n^2 + 3n}{5n^2 + 7} \quad \implies \quad \lim_{n \to \infty} a_n = \frac{2}{5}.$

#### B. **Comparison with Known Limits**
- Compare with simpler sequences whose limits are known.

#### C. **L'HÃ´pital's Rule** (for indeterminate forms like $\frac{0}{0}$ or $\frac{\infty}{\infty}$:
$\lim_{n \to \infty} \frac{f(n)}{g(n)} = \lim_{n \to \infty} \frac{f'(n)}{g'(n)} \quad \text{if derivatives exist.}$

---

#Limits of Real Functions
---

### **1. Definition of the Limit of a Real Function**

Let $f(x)$ be a real-valued function defined on some domain $D \subset \mathbb{R}$.  
The **limit** of $f(x)$ as $x \to c$ is $L$ if, for every $\epsilon > 0$, there exists a $\delta > 0$ such that:  
$|f(x) - L| < \epsilon \quad \text{whenever } 0 < |x - c| < \delta.$
- Denoted as:
  $$\lim_{x \to c} f(x) = L.$$
- $L$: The limit of $f(x)$ as $x$ approaches $c$.  

If no such $L$ exists, the limit does not exist.

---

### **2. One-Sided Limits**
#### A. **Right-Hand Limit**
$$\lim_{x \to c^+} f(x) = L \quad \text{if } f(x) \to L \text{ as } x \to c \text{ from the right } (x > c).$$

#### B. **Left-Hand Limit**
$$\lim_{x \to c^-} f(x) = L \quad \text{if } f(x) \to L \text{ as } x \to c \text{ from the left } (x < c).$$

#### C. Existence of Two-Sided Limit
The two-sided limit exists if and only if:  
$$\lim_{x \to c^+} f(x) = \lim_{x \to c^-} f(x) = L.$$

---

### **3. Limit at Infinity**
#### A. Finite Limit as \( x \to \infty \)
$$\lim_{x \to \infty} f(x) = L \quad \text{if for every } \epsilon > 0, \text{ there exists } M > 0 \text{ such that } |f(x) - L| < \epsilon \text{ for all } x > M.$$

#### B. Infinite Limit
- $$f(x) \to \infty$:  
  $\lim_{x \to c} f(x) = \infty \quad \text{if } f(x) \text{ grows arbitrarily large as } x \to c.$$

---

### **4. Properties of Limits**

#### A. Sum Rule
$$\lim_{x \to c} [f(x) + g(x)] = \lim_{x \to c} f(x) + \lim_{x \to c} g(x).$$

#### B. Difference Rule
$$\lim_{x \to c} [f(x) - g(x)] = \lim_{x \to c} f(x) - \lim_{x \to c} g(x).$$

#### C. Product Rule
$$\lim_{x \to c} [f(x) \cdot g(x)] = \left(\lim_{x \to c} f(x)\right) \cdot \left(\lim_{x \to c} g(x)\right).$$

#### D. Quotient Rule
$$\lim_{x \to c} \frac{f(x)}{g(x)} = \frac{\lim_{x \to c} f(x)}{\lim_{x \to c} g(x)}, \quad \text{if } \lim_{x \to c} g(x) \neq 0.$$

#### E. Power Rule
$$\lim_{x \to c} [f(x)]^n = \left[\lim_{x \to c} f(x)\right]^n, \quad n \in \mathbb{N}.$$

---

### **5. Special Cases**
#### A. Polynomial Functions
$$\lim_{x \to c} P(x) = P(c),$$
where $P(x)$ is a polynomial.

#### B. Rational Functions
$$\lim_{x \to c} \frac{P(x)}{Q(x)} = \frac{P(c)}{Q(c)}, \quad Q(c) \neq 0.$$
If $Q(c) = 0$, use limits at the boundary (or simplify).

#### C. Trigonometric Limits
1. $$\lim_{x \to 0} \frac{\sin x}{x} = 1$$.
2. $$\lim_{x \to 0} \frac{1 - \cos x}{x} = 0$$.

#### D. Exponential and Logarithmic Limits
1. $$\lim_{x \to \infty} e^x = \infty, \lim_{x \to -\infty} e^x = 0 $$.
2. $$\lim_{x \to 0^+} \ln x = -\infty, \lim_{x \to \infty} \ln x = \infty$$.

---

### **6. Techniques for Finding Limits**

#### A. **Direct Substitution**
If $f(x)$ is continuous at $c$, then:  
$$\lim_{x \to c} f(x) = f(c).$$

#### B. **Factoring**
Simplify $f(x)$ by factoring and canceling terms.

#### C. **Rationalizing**
Multiply numerator and denominator by a conjugate to simplify.

#### D. **Divide by the Highest Power**
For limits involving rational functions as $x \to \infty$:  
- Divide numerator and denominator by the highest power of $x$.

#### E. **Lâ€™HÃ´pitalâ€™s Rule** (for indeterminate forms like $\frac{0}{0}$ or $\frac{\infty}{\infty}$:
$$\lim_{x \to c} \frac{f(x)}{g(x)} = \lim_{x \to c} \frac{f'(x)}{g'(x)}, \quad \text{if derivatives exist.}$$

---

### **7. Squeeze Theorem**
If $g(x) \leq f(x) \leq h(x)$ for all $x$ near $c$, and:
$$\lim_{x \to c} g(x) = \lim_{x \to c} h(x) = L,$$
then:
$$\lim_{x \to c} f(x) = L.$$
- Example:  
  $$\lim_{x \to 0} x^2 \sin \frac{1}{x} = 0, \quad \text{using } -x^2 \leq x^2 \sin \frac{1}{x} \leq x^2.$$
---

### **8. Infinite Limits and Asymptotes**

#### A. Vertical Asymptotes
If $$\lim_{x \to c^+} f(x) = \pm \infty$$ or $$\lim_{x \to c^-} f(x) = \pm \infty$$
then $x = c$ is a vertical asymptote.

#### B. Horizontal Asymptotes
If $$\lim_{x \to \infty} f(x) = L$$, then $y = L$ is a horizontal asymptote.
