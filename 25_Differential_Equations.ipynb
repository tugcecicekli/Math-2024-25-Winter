### Notes on Differential Equations

#### 1. **What is a Differential Equation?**
A **differential equation** is an equation that involves an unknown function and its derivatives. These equations describe the relationship between a function and the rates at which it changes.

- **Ordinary Differential Equation (ODE)**: Involves derivatives with respect to a single variable (usually time).
- **Partial Differential Equation (PDE)**: Involves partial derivatives with respect to more than one variable (e.g., time and space).

#### 2. **Types of Differential Equations**

- **Order**: The highest derivative in the equation.
  - *First-order* equations involve the first derivative.
  - *Second-order* equations involve the second derivative, and so on.
  
- **Linearity**:
  - A differential equation is **linear** if the unknown function and its derivatives appear to the power of one (i.e., no products or powers of the function or its derivatives).
  - **Nonlinear** if there are products, powers, or other nonlinear terms involving the function or its derivatives.

- **Homogeneous vs Non-homogeneous**:
  - A differential equation is **homogeneous** if every term involves the function or its derivatives.
  - It is **non-homogeneous** if there is an additional non-zero term (often referred to as the forcing term or external function).

#### 3. **General Solution & Particular Solution**
- **General Solution**: The most general form of the solution, typically involving arbitrary constants.
- **Particular Solution**: A solution obtained by substituting specific values for constants based on initial/boundary conditions.

#### 4. **Methods of Solving Differential Equations**

- **Separation of Variables**: For separable equations of the form $\frac{dy}{dx} = g(x)h(y)$, you can separate the variables and integrate both sides.
  
- **Integrating Factor**: Used to solve linear first-order ODEs of the form $\frac{dy}{dx} + P(x)y = Q(x)$. Multiply the equation by an integrating factor to make the left-hand side an exact derivative.

- **Homogeneous Equations**: For higher-order linear ODEs, solving homogeneous equations often involves finding characteristic equations.

- **Characteristic Equation (Linear ODEs)**: For linear differential equations with constant coefficients, the characteristic equation is obtained by assuming a solution of the form $y = e^{rx}$, leading to a polynomial equation for $r$.

- **Variation of Parameters**: Used for non-homogeneous linear ODEs, this method involves using the solution of the homogeneous equation and adjusting parameters to account for the non-homogeneous part.

- **Laplace Transform**: A powerful tool for solving linear ODEs, especially with initial conditions. It transforms the ODE into an algebraic equation in the complex $s$-domain, which is easier to solve.

- **Numerical Methods**: When analytical solutions are difficult or impossible, methods like Euler‚Äôs method, Runge-Kutta methods, or finite difference methods can be used to approximate solutions.

#### 5. **Example of a First-Order ODE (Separation of Variables)**:

   - $\frac{dy}{dx} = \frac{y}{x} \quad \text{(Separable)}$
   Step-by-step solution:
   - Rewrite as $\frac{dy}{y} = \frac{dx}{x}$.
   - Integrate both sides: $\int \frac{dy}{y} = \int \frac{dx}{x}$, yielding:
   $\ln|y| = \ln|x| + C$
   - Exponentiate both sides to solve for $y$:
   $y = Cx$

#### 6. **Second-Order Linear Differential Equations**
   A general second-order linear ODE:
   $a(x) \frac{d^2y}{dx^2} + b(x) \frac{dy}{dx} + c(x) y = f(x)$

   Where $a(x)$, $b(x)$, and $c(x)$ are functions of $x$, and $f(x)$ is a given function.
   - **Homogeneous Case** $f(x) = 0$:

     $a(x) \frac{d^2y}{dx^2} + b(x) \frac{dy}{dx} + c(x) y = 0$
     - Solutions typically involve characteristic equations for constant coefficients.
   
   - **Non-Homogeneous Case** $f(x) \neq 0$:
     Use **method of undetermined coefficients** or **variation of parameters** for specific solutions.

#### 7. **Systems of Differential Equations**
   - **Coupled Equations**: A system of two or more equations involving multiple dependent variables.
     - Example:
       $\frac{dx}{dt} = y, \quad \frac{dy}{dt} = -x$
     - The system describes a harmonic oscillator, which has sinusoidal solutions.

#### 8. **Partial Differential Equations (PDEs)**

- **Form**: A PDE involves partial derivatives of an unknown function with respect to more than one independent variable.
  - Example: Heat equation, wave equation, Laplace equation.

- **Solution Methods**:
  - **Separation of Variables**: Separate the spatial and time variables.
  - **Fourier Series/Transforms**: Decompose the solution into sinusoidal components.
  - **Numerical Methods**: Finite difference, finite element, and other discretization techniques.
